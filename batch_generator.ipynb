```python
import numpy as np

def generate_batches(X, y, batch_size):
    """
    param X: np.array[n_objects, n_features]
    param y: np.array[n_objects]
    """
    assert len(X) == len(y)
    np.random.seed(42)
    X = np.array(X)
    y = np.array(y)
    perm = np.random.permutation(len(X))
    ind = []
    start = 0
    end = batch_size

    for batch_start in range(len(X)//batch_size):
      ind = perm[start:end]
      yield (X[np.array(ind)], y[np.array(ind)])
      ind = []
      start += batch_size
      end += batch_size
```

```python
X_fake = np.arange(100)[:, np.newaxis]
y_fake = np.arange(100) + 1000
```

```python
k = 0
for batch_X, batch_y in generate_batches(X_fake, y_fake, 9):
  print("Batch №", k, "\n", batch_X,batch_y, "\n\n")
  k += 1
```
